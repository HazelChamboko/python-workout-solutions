To solve this, I used the itemgetter. This is a function in the= operator module. PS. The operator module is one of the coolest so do not forget to check it out!

```python
from operator import itemgetter

get_item = itemgetter(1, -1)
get_item(['mary',' jose','aaron'])
```

```python
from operator import itemgetter

def get_i(*items):
    output = itemgetter(0, -1)
    for i in items:
        return output(*items)
print(get_i([1,2,3,4,5,6]))
print(get_i(['mary', 'eliza', 'zac','Vee']))    
```

```python
def mysum(*items):
    if not items: 
        return items
    output = items[0]
    for item in items[1:]:
        output += item  
    return output

print(mysum())
print(mysum(10, 20, 30, 40))
print(mysum('a', 'b', 'c', 'd'))
print(mysum([10, 20, 30], [40, 50, 60], [70, 80]))

```

```python
import operator

PEOPLE = [{'first': 'Reuven', 'last': 'Lerner',
           'email': 'reuven@lerner.co.il'},
          {'first': 'Donald', 'last': 'Trump',
           'email': 'president@whitehouse.gov'},
          {'first': 'Vladimir', 'last': 'Putin',
           'email': 'president@kremvax.ru'}
          ]

def alphabetize_names(list_of_dicts):
    return sorted(list_of_dicts,
        key=operator.itemgetter('last', 'first'))

print(alphabetize_names(PEOPLE))
```

```python
from collections import Counter
import operator

WORDS = ['this', 'is', 'an',
         'elementary', 'test', 'example']

def most_repeating_letter_count(word): 
    return Counter(word).most_common(1)[0][1] 


def most_repeating_word(words):
    return max(words, key=most_repeating_letter_count) 

               
print(most_repeating_word(WORDS))
```
