This program is an exercise from the Pyhton Workout, its called a pig latin, a language used in American slang, whenever a word starts with a vowel, a "way" is added at the end of the word and when a non vowel starts the first letter of the word goes to the end of the word and an "ay" is added as shown in the code. I have learnt that strings can use slicing methods and indexing methods for manipulation.

```python
def pl_sentence(sentence):
    output = []
    for word in sentence.split():
        if word[0] in 'aeiou':
            output.append(f'{word}way')
        else:
            output.append(f'{word[1:]}{word[0]}ay') 
    return ' '.join(output)  

pl_sentence("i jumped the big fox once")        
             
```

## EXERCISE 7: UBBI DUBBI

```python
def ubbi_dubbi(sentence):
    output = []
    for word in sentence:
        if word in 'aeiou':
            output.append(f'ub{word}')
        else:
            output.append(word)
            
    return ''.join(output)
print(ubbi_dubbi("moon"))        
        
```
